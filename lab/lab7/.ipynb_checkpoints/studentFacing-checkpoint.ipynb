The original paper can be found [here](http://statmodeling.stat.columbia.edu/wp-content/uploads/2016/04/transphobia-and-canvassing.pdf). Existing research depicts intergroup prejudices as deeply ingrained, requiring intense intervention to lastingly reduce. Here, we show that a single approximately 10-minute conversation encouraging actively taking the perspective of others can markedly reduce prejudice for at least 3 months. A randomized trial found that these conversations substantially reduced transphobia, with decreases greater than Americansâ€™ average decrease in homophobia from 1998 to 2012. These effects persisted for 3 months, and both transgender and nontransgender canvassers were effective. The interventio nalso increased support for a nondiscrimination law, even after exposing voters to counterarguments

## ```seaborn``` and ```matplotlib```

[```seaborn```](https://seaborn.pydata.org/) is a powerful visualization package built upon another package called [```matplotlib```](https://matplotlib.org/). Let's take a look at these two packages in action!

Run the next cell to read in a demo dataframe.

````python
tips = sns.load_dataset("tips")
tips.head()
````

To create a simple scatterplot using ```seaborn```, we'll use the ```sns.scatterplot``` function.

This function uses the following arguments:
 * ```data```: The dataframe that you are using in this plot
 * ```x```: The column of the dataframe to use as the x coordinates of the points
 * ```y```: The column of the dataframe to use as the y coordinates of the points
 * ```s```: The size of the data points
 
Make sure that you put the name of each column in quotes. 'Single quotes' or "Double quotes" are both okay.

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", s=100)
````

Pretty cool, right? What about the ```<matplotlib.axes._subplots.AxesSubplot at 0x195892e6188>``` at the top of the graph? To get rid of the extra text, add in a ```;``` to the last line of code. Run the next cell to see how this works.

````python
sns.scatterplot(data=tips, x="total_bill", y="tip");
````

What if we want to see which data points in out scatterplot correspond to which meal? You can set the ```hue``` argument to variable ```"time``` to have the "Lunch" data points be set to blue and the "Dinner" data points to be set to orange.

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time");
````

Let's try adding in a helpful title, x-axis label, and y-axis label.

**Notice that to add a title, x label, and y label, use ```plt.``` instead of ```sns.```**

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Lunch Tips vs Dinner Tips")
plt.ylabel("Tip Amount, in Dollars")
plt.xlabel("Total Bill Amount, in Dollars");
````

What if you want to add other elements to the plot? Let's add some text, lines, and a legend to the plot.

Here are the arguments of the functions in use below:
 * ```plt.plot```:
  * [10, 10]: The first list is the starting and ending **x value**. Since we want a vertical line, these numbers should be the same
  * [2, 10]: The second list is the starting and ending **y value**. In this example, we want the line to start at y = 2 and end at y = 10
  * color: This allows us to select what color we would like to have the line be
  * linestyle: This allows us to select if we want a solid line, dashed line, or other type of line
 * ```plt.text```:
  * s: This is the text you would like to have on the plot
  * x: This is the x value where the text should start
  * y: This is the y value where the text should start
  * rotation: This value rotates the text

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")

plt.plot([10,10], [2,10], color = "black", linestyle = "dashed")
plt.text(s = "$10", x = 8, y = 6)
plt.text(s = "$10", x = 8, y = 8, rotation = 90)

plt.title("Lunch Tips vs Dinner Tips")
plt.ylabel("Tip Amount, in Dollars")
plt.xlabel("Total Bill Amount, in Dollars");
````

The following cell does the following:
 * ```df[["se.low","se.high"]]```: Selects the ```se.low``` and ```se.high``` columns from the ```df``` dataframe
 * ```.values```: Gets the values for each row
 * ```.tolist()```: Puts the values into a list
