### The Ultimatum Game

Previous studies have shown that humans typically reject low offers in bargaining games, even when doing so goes against their material self-interest.

In these “ultimatum games”, a proposer makes an offer to a responder for how to divide a fixed prize. A responder then decides whether to accept or reject the offer. 

If it is accepted, both players divide the prize as agreed. 

If it is rejected, both players receive nothing. If both players are completely rational and selfish—an assumption that is widely used by social scientists in formal models of inter- national bargaining —then proposers will offer almost nothing to responders, an offer that rational responders nonetheless accept because something is better than nothing.
