```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
topinc10 = pd.read_excel("data/TopIncomesPanel.xls", sheet_name="ExCG_Top10_Adj",skiprows=1)
topinc10
```

```python
topinc1 = pd.read_excel("data/TopIncomesPanel.xls", sheet_name="ExCG_Top1_Adj",skiprows=1)
topinc1
```

```python
topinc1 = topinc1.rename({"Calendar Year":"Year"}, axis=1)
topinc10 = topinc10.rename({"Calendar Year":"Year"}, axis=1)
```

```python
topinc1long = pd.melt(topinc1,id_vars="Year", value_vars= ['Australia', 'Canada', 'France', 'Germany', 'Ireland', 'Japan',
       'Netherlands', 'New Zealand', 'Spain', 'Sweden', 'Switzerland', 'UK',
       'US'], var_name="Country", value_name="Top1inc")
```

```python
topinc10long = pd.melt(topinc1,id_vars="Year", value_vars= ['Australia', 'Canada', 'France', 'Germany', 'Ireland', 'Japan',
       'Netherlands', 'New Zealand', 'Spain', 'Sweden', 'Switzerland', 'UK',
       'US'], var_name="Country", value_name="Top10inc")
```

```python
ss = pd.read_csv("data/Scheve_Stasavage_IO_2010_CoWreplicationdata.csv")
ss
```

```python
ss.country[ss.country=="USA"] = "US"
```

```python
ss
```

```python
ss=ss.rename({"country":"Country", "year":"Year"}, axis=1)
```

```python
ssplus  = pd.merge(ss, topinc1long, on=["Country", "Year"], how="left")
ssplus
```

```python
ssplus  = pd.merge(ssplus, topinc10long, on=["Country", "Year"], how="left")
ssplus
```

```python
ssclose = ssplus[(ssplus.Year >= 1900)]
ssclose = ssclose[ssclose.Year <= 1930]
```

```python
smf.ols("Top1inc ~ wwihighmobaft + Year + Country", data=ssclose).fit().summary()
```
